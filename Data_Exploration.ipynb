# <span style='color:black'>Import libraries</span>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# <span style='color:black'>Load Iris dataset</span>

```python
# Load Iris dataset
iris = load_iris()
```

# <span style='color:black'>Convert to Dataframe</span>

```python
# Convert to DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

# <span style='color:black'>Explore dataset structure</span>

```python
# Explore dataset structure
print("Dataset structure:")
print(iris_df.info())
```

# <span style='color:black'>Check for missing values</span>

```python
# Check for missing values
print("\nMissing values:")
print(iris_df.isnull().sum())
```

# <span style='color:black'>Visualize the data using plots and graphs</span>

# <span style='color:black'>Pairplot to visualize pairwise relationships in the dataset</span>

```python
# Visualize the data using plots and graphs
# Pairplot to visualize pairwise relationships in the dataset
sns.pairplot(iris_df, hue='target', palette='viridis')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

# <span style='color:black'>Boxplot to visualize the distribution of each feature</span>

```python
# Boxplot to visualize the distribution of each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop(columns='target'))
plt.title('Boxplot of Iris Dataset Features')
plt.show()
```

# <span style='color:black'>Violinplot to visualize the distribution of each feature</span>

```python
# Violinplot to visualize the distribution of each feature
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris_df.drop(columns='target'))
plt.title('Violinplot of Iris Dataset Features')
plt.show()
```
